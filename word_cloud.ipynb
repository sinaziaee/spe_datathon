```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the dataset (modify the file path as needed)
file_path = 'C:\\datasets\\spe\\regulations.xlsx'
data = pd.read_excel(file_path)

# Combine the relevant columns to create a text for the word cloud
text_data = data['Regulation'].astype(str) + " " + data['Region/Country'].astype(str) + " " + data['Focus Area'].astype(str) + " " + data['Description'].astype(str)

# Join all text data into a single string
combined_text = " ".join(text_data)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(combined_text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
